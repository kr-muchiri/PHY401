```python
def phase_oracle():
    qc = QuantumCircuit(2)
    qc.cz(0,1)  # Introduces phase flip for |1⟩ state
    return qc

oracle = phase_oracle()
oracle.draw('mpl')
```

```python
def initial_state():
    qc = QuantumCircuit(2)
    qc.h([0,1])  # Place both qubits in superposition
    qc.z(1)  # Convert |+⟩ to |-⟩ for the auxiliary qubit
    return qc

init_state = initial_state()
init_state.draw('mpl')
```

```python
def full_circuit():
    qc = QuantumCircuit(2, 2)
    qc += initial_state()
    qc += phase_oracle()
    return qc

qc = full_circuit()
qc.draw('mpl')
```

```python
from qiskit import QuantumCircuit, Aer, transpile, assemble
from qiskit.visualization import plot_histogram

# Create a quantum circuit with 2 qubits and 1 classical bit
qc = QuantumCircuit(2, 1)

# Initialize the qubits
qc.x(1)

# Apply H gates to both qubits
qc.h([0,1])

# Oracle: For this example, let's use an Oracle for a balanced function
# You may need to change this part to represent the specific function you have in mind.
qc.cx(0,1)

# Apply H gate to the first qubit
qc.h(0)

# Measure the first qubit
qc.measure(0, 0)

# Visualize the circuit
qc.draw('mpl')

# Simulate the circuit
simulator = Aer.get_backend('qasm_simulator')
compiled_circuit = transpile(qc, simulator)
qobj = assemble(compiled_circuit)
result = simulator.run(qobj).result()

# Get the results and plot them
counts = result.get_counts()
plot_histogram(counts)
```
