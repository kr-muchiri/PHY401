```python
def classical_factorization(n):
    """ Factorize a number using classical trial division method """
    factors = []
    # Check for factors 2 separately to make the algorithm more efficient
    while n % 2 == 0:
        factors.append(2)
        n //= 2
    
    # Check for odd factors starting from 3
    factor = 3
    max_factor = n**0.5
    while factor <= max_factor:
        while n % factor == 0:
            factors.append(factor)
            n //= factor
            max_factor = n**0.5
        factor += 2

    # If n is a prime number greater than 2
    if n > 1:
        factors.append(n)
    
    return factors

# Example usage
number = 91
factors = classical_factorization(number)
print(f"Factors of {number}: {factors}")
```

```python
import matplotlib.pyplot as plt

def classical_factorization_visualized(n):
    factors = []
    counts = {}
    # Check for factors 2 separately
    while n % 2 == 0:
        factors.append(2)
        n //= 2
        counts[2] = counts.get(2, 0) + 1
    
    # Check for odd factors
    factor = 3
    while factor <= n:
        while n % factor == 0:
            factors.append(factor)
            n //= factor
            counts[factor] = counts.get(factor, 0) + 1
        factor += 2

    # Visualization
    plt.bar(counts.keys(), counts.values())
    plt.xlabel('Divisors')
    plt.ylabel('Counts')
    plt.title(f'Factorization of {number}')
    plt.show()

    return factors

# Example usage
number = 91
classical_factorization_visualized(number)
```
